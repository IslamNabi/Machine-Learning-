```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

# Generate synthetic data that is not perfectly separable
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, 
                           n_clusters_per_class=1, flip_y=0.1, random_state=42)  # flip_y adds noise
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.bwr, edgecolors='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Binary Classification Data')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Fit linear regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Predict using linear regression
y_pred_lin = lin_reg.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred_lin, cmap=plt.cm.bwr, edgecolors='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Linear Regression Predictions (Continuous Values)')
plt.colorbar(label='Predicted Value')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X, y)

# Predict using logistic regression
y_pred_log = log_reg.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_pred_log, cmap=plt.cm.bwr, edgecolors='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Predictions (Class Labels)')
plt.show()
```

```python
# Create a grid of points
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

# Predict using linear regression
Z_lin = lin_reg.predict(np.c_[xx.ravel(), yy.ravel()])
Z_lin = Z_lin.reshape(xx.shape)

# Plot decision boundary
plt.contourf(xx, yy, Z_lin, alpha=0.8, cmap=plt.cm.bwr)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.bwr, edgecolors='k')
plt.title('Linear Regression Decision Boundary')
plt.show()
```

```python
# Predict using logistic regression
Z_log = log_reg.predict(np.c_[xx.ravel(), yy.ravel()])
Z_log = Z_log.reshape(xx.shape)

# Plot decision boundary
plt.contourf(xx, yy, Z_log, alpha=0.8, cmap=plt.cm.bwr)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.bwr, edgecolors='k')
plt.title('Logistic Regression Decision Boundary')
plt.show()
```
