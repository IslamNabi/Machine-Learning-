# Uni Variate Analysis

```python
# Setting figure size
plt.rcParams['figure.figsize'] = (10, 4)

# Plotting the Distribution of customers on Revenue (buy or not)
plt.subplot(1, 2, 1)
sns.countplot(x='Revenue', data=data, palette='pastel')
plt.title('Buy or Not', fontsize=20)
plt.xlabel('Revenue or not', fontsize=15)
plt.ylabel('count', fontsize=15)

# Plotting the Distribution of customers on Weekend (purchase on weekends or not)
plt.subplot(1, 2, 2)
sns.countplot(x='Weekend', data=data, palette='pastel')
plt.title('Purchase on Weekends', fontsize=20)
plt.xlabel('Weekend or not', fontsize=15)
plt.ylabel('count', fontsize=15)

# Display the plots
plt.show()


```

## Plotting a Pie Chart for Browsers 

```python
plt.rcParams['figure.figsize'] = (20, 10)
colors_visitors = sns.color_palette("pastel")[0:3]  # First 3 colors from Seaborn's pastel palette

size_visitors = [10551, 1694, 85]
labels_visitors = ["Returning Visitor", "New Visitor", "Others"]
explode_visitors = [0, 0, 0.1]  # Highlight "Others" with a small separation

# Plotting the first pie chart (Visitors)
plt.subplot(1, 2, 1)
plt.pie(size_visitors, colors=colors_visitors, labels=labels_visitors, explode=explode_visitors, 
         autopct='%.2f%%', wedgeprops={'edgecolor': 'black'})
plt.title('Different Visitors', fontsize=30)
plt.axis('off')
plt.legend()

# Data for browsers pie chart
size_browsers = [7961, 2462, 736, 467, 174, 163, 300]
labels_browsers = ["2", "1", "4", "5", "6", "10", "Others"]

colors_browsers = sns.color_palette("Set2")  # Colors from Seaborn's Set2 palette

# Plotting the second pie chart (Browsers)
plt.subplot(1, 2, 2)
plt.pie(size_browsers, colors=colors_browsers, labels=labels_browsers, 
       autopct='%.2f%%', startangle=90, wedgeprops={'edgecolor': 'black'})
plt.title('Different Browsers', fontsize=30)
plt.axis('off')
plt.legend()

# Displaying the plots
plt.show()
```

## visualizing the distribution of customers around the Region

# Bi Variate Analysis

## Analyzing a relationship between Revenue and another variable.

```python
plt.rcParams['figure.figsize'] = (14, 10)

# Informational Duration vs Revenue
plt.subplot(2, 2, 1)
sns.boxenplot(x='Revenue', y='Informational_Duration', data=data, palette='rainbow')
plt.title('Info. duration vs Revenue', fontsize=15)
plt.xlabel('Revenue', fontsize=15)
plt.ylabel('Informational Duration', fontsize=15)

# Administrative Duration vs Revenue
plt.subplot(2, 2, 2)
sns.boxenplot(x='Revenue', y='Administrative_Duration', data=data, palette='pastel')
plt.title('Admn. duration vs Revenue', fontsize=15)
plt.xlabel('Revenue', fontsize=15)
plt.ylabel('Administrative Duration', fontsize=15)

# Product Related Duration vs Revenue
plt.subplot(2, 2, 3)
sns.boxenplot(x='Revenue', y='ProductRelated_Duration', data=data, palette='dark')
plt.title('Product Related duration vs Revenue', fontsize=15)
plt.xlabel('Revenue', fontsize=15)
plt.ylabel('Product Related Duration', fontsize=15)

# Exit Rate vs Revenue
plt.subplot(2, 2, 4)
sns.boxenplot(x='Revenue', y='ExitRates', data=data, palette='spring')
plt.title('ExitRates vs Revenue', fontsize=15)
plt.xlabel('Revenue', fontsize=15)
plt.ylabel('ExitRates', fontsize=15)

plt.show()
```

## Page values vs Revenue and Bounce rates vs Revenue

```python
plt.rcParams['figure.figsize'] = (14, 5)

plt.subplot(1, 2, 1)
sns.stripplot(x='Revenue', y='PageValues', data=data, palette = 'autumn')
plt.title('PageValues vs Revenue', fontsize = 30)
plt.xlabel('PageValues', fontsize = 15)
plt.ylabel('Revenue', fontsize = 15)

plt.subplot(1, 2, 2)
sns.stripplot(x='Revenue', y='BounceRates', data=data, palette = 'magma')
plt.title('Bounce Rates vs Revenue', fontsize = 30)
plt.xlabel('Boune Rates', fontsize = 15)
plt.ylabel('Revenue', fontsize = 15)

plt.show()
```
