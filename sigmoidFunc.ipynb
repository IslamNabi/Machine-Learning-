```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# Test with a few values
print(sigmoid(0))  # Should be 0.5
print(sigmoid(10))  # Should be close to 1
print(sigmoid(-10))  # Should be close to 0
```

```python
# Create a range of values for z
z = np.linspace(-10, 10, 100)  # 100 points between -10 and 10

# Calculate the sigmoid for each z
sigmoid_values = sigmoid(z)

# Plot the sigmoid function
plt.plot(z, sigmoid_values)
plt.xlabel('z')
plt.ylabel('Sigmoid(z)')
plt.title('Sigmoid Function')
plt.grid(True)
plt.show()
```
