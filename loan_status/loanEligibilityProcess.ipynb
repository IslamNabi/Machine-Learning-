```python
import pandas as pd

data = pd.read_csv('loan_data_set.csv')
data
```

```python
data.shape
```

```python
print(data.isnull().sum())
```

```python
data_cleaned = data.dropna()
```

```python
print(data_cleaned.shape)
```

```python
data_cleaned['Loan_Status'] = data_cleaned['Loan_Status'].replace({'Y': 1, 'N': 0})
print(data_cleaned['Loan_Status'].head())
```

```python
data_cleaned['Gender'] = data_cleaned['Gender'].replace({'Male': 1, 'Female': 0})
print(data_cleaned['Gender'].head())
```

```python
data_cleaned['Married'] = data_cleaned['Married'].replace({'Yes': 1, 'No': 0})
print(data_cleaned['Married'].head())
```

```python
data_cleaned['Education'] = data_cleaned['Education'].replace({'Graduate': 1, 'Not Graduate': 0})
print(data_cleaned['Education'].head())
```

```python
data_cleaned['Self_Employed'] = data_cleaned['Self_Employed'].replace({'Yes': 1, 'No': 0})
print(data_cleaned['Self_Employed'].head())
```

```python
X = data_cleaned.drop(columns=['Loan_Status'])
X
```

```python
y = data_cleaned['Loan_Status']
y
```

```python
data_cleaned['Property_Area'] = data_cleaned['Property_Area'].replace({'Rural': 0, 'Urban': 1, 'Semiurban': 2})
print(data_cleaned['Property_Area'].head())
```

```python
X = data_cleaned.drop(columns=['Loan_ID', 'Dependents'])
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Train the Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(cm)
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print('Classification Report:')
print(report)
```
