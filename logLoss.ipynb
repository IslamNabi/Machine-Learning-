```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost functions
def squared_error_cost(y_true, y_pred):
    return (y_true - y_pred) ** 2

def logistic_loss(y_true, y_pred):
    # Clip predictions to avoid log(0) error
    y_pred = np.clip(y_pred, 1e-10, 1 - 1e-10)
    return - (y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# True label
y_true = 1

# Predicted probabilities (from 0.01 to 0.99)
y_pred = np.linspace(0.01, 0.99, 100)

# Calculate costs
squared_error = squared_error_cost(y_true, y_pred)
log_loss = logistic_loss(y_true, y_pred)

# Plot the cost functions
plt.plot(y_pred, squared_error, label='Squared Error Cost', linestyle='dashed', color='blue')  # U-shaped curve
plt.plot(y_pred, log_loss, label='Logistic Loss', linestyle='solid', color='red')  # Asymmetric curve
plt.xlabel('Predicted Probability (y_pred)')
plt.ylabel('Cost')
plt.title('Squared Error Cost vs Logistic Loss')
plt.legend()
plt.grid(True)
plt.show()
```
